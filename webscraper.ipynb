```python
import requests
import html5lib
import string
import pandas as pd
from bs4 import BeautifulSoup, Tag
```

```python
url = "https://www.vlr.gg/event/stats/1188/champions-tour-2023-lock-in-s-o-paulo?exclude=&min_rounds=0&agent=all"

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36'
}
```

```python
r = requests.get(url, {'headers': headers})
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find_all('table')[0]

table
```

```python
agents = table.find_all("td", {"class": "mod-agents"})

new_agents = []
for td in agents:
    div = td.find('div')
    new_td = Tag(name='td', attrs={'class': 'mod-agents'})
    for img in div.find_all('img'):
        new_p = Tag(name='p')
        new_p.string = f'{img["src"].split("/")[-1].split(".")[0].title()} '
        new_td.append(new_p)
    new_agents.append(new_td)
    
for i, td in enumerate(table.find_all("td", {"class": "mod-agents"})):
    td.replace_with(new_agents[i])
    
table.find_all("td", {"class": "mod-agents"})
```

```python
df = pd.read_html(str(table))[0]
df
```

```python
titles = ["Players"]
for th in table.find_all("th"):
    title = th.get("title")
    if title:
        titles.append(title)
        
df = df.set_axis(titles, axis=1, copy=False)
df
```

```python
def percentage_to_float(column):
    return column.str.rstrip('%').astype('float') / 100

df['Clutch Success %'] = percentage_to_float(df['Clutch Success %'])
df['Kill, Assist, Trade, Survive %'] = percentage_to_float(df['Kill, Assist, Trade, Survive %'])
df['Headshot %'] = percentage_to_float(df['Headshot %'])
df
```

```python
df[['Cluthes Won', 'Clutches Played']] = df['Clutches (won/played)'].str.split('/', expand=True)
df['Clutches Played'] = df['Clutches Played'].fillna(0)
df
```

```python
df.drop(['Clutches (won/played)'], axis='columns', inplace=True)
df
```
